```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
df['country.name']= label_encoder.fit_transform(df['country.name'])
```

```python
df['doctypebranch']= label_encoder.fit_transform(df['doctypebranch'])
```

```python
df['originatingbody_type']= label_encoder.fit_transform(df['originatingbody_type'])
```

```python
df['respondent.0']= label_encoder.fit_transform(df['respondent.0'])
df['respondentOrderEng']= label_encoder.fit_transform(df['respondentOrderEng'])
df=df.drop(['respondent.1','respondent.2','respondent.3','respondent.4'],axis=1)
```

```python
df['originatingbody_name']= label_encoder.fit_transform(df['originatingbody_name'])
```

```python
df=df.drop(['parties.2','parties.1','parties.0'],axis=1)
```

```python
X = df[['appno']]
df=df.drop(['appno'],axis=1)
```

```python
Y = df[['importance']]
df=df.drop(['importance'],axis=1)
```

```python
# df['kpdate'] = pd.to_datetime(df['kpdate'])
# df = df.assign(
#                day_k=df.kpdate.dt.day,
#                month_k=df.kpdate.dt.month,
#                year_k=df.kpdate.dt.year)

# df = df.drop(['kpdate'],axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df,Y,random_state=0,test_size=.2)
```

```python
l = [X for X in X_train.columns]
len(l)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=1,n_estimators=100)
```

```python
rf.fit(X_train,y_train)
```

```python
predict_rf=rf.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
```

```python
print(mean_absolute_error(predict_rf,y_test))
print(accuracy_score(predict_rf,y_test))
```
